In this exercise we will implement fully-connected networks using a modular approach. For each layer we will implement a `forward` and a `backward` function. The `forward` function will receive inputs, weights, and other parameters and will return both an output and a `cache` object storing data needed for the backward pass, like this:

```python

```

```python

  return dx, dw
```

After implementing a bunch of layers this way, we will be able to easily combine them to build classifiers with different architectures.

````python
# Test the affine_forward function

num_inputs = 2
input_shape = (4, 5, 6)
output_dim = 3

input_size = num_inputs * np.prod(input_shape)
weight_size = output_dim * np.prod(input_shape)

x = np.linspace(-0.1, 0.5, num=input_size).reshape(num_inputs, *input_shape)
w = np.linspace(-0.2, 0.3, num=weight_size).reshape(np.prod(input_shape), output_dim)
b = np.linspace(-0.3, 0.1, num=output_dim)

'''
np.linspace(-0.1, 0.5, num=input_size)：生成一个长度为 input_size 的一维数组，
其中的元素从 -0.1 到 0.5 均匀分布。这个数组被用作输入数据 x。
reshape(num_inputs, *input_shape)：将长度为 num_inputs 的一维数组 x 转换为一个 num_inputs 行、以 input_shape 为形状的多维数组。
这里使用了 * 运算符，将 input_shape 这个元组解包成了三个参数，
相当于 reshape(num_inputs, input_shape[0], input_shape[1], input_shape[2])。

np.linspace(-0.2, 0.3, num=weight_size)：生成一个长度为 weight_size 的一维数组，其中的元素从 -0.2 到 0.3 均匀分布。这个数组被用作神经网络的权重矩阵 w。
reshape(np.prod(input_shape), output_dim)：将长度为 np.prod(input_shape) 的一维数组 w 转换为一个 np.prod(input_shape) 行、
output_dim 列的二维数组。其中 np.prod(input_shape) 表示 input_shape 中所有元素的乘积，也就是 input_shape 所代表的向量的长度。
这个数组表示神经网络每个输入特征与每个输出特征之间的权重。
np.linspace(-0.3, 0.1, num=output_dim)：生成一个长度为 output_dim 的一维数组，
其中的元素从 -0.3 到 0.1 均匀分布。这个数组被用作神经网络的偏置向量 b。
'''

out, _ = affine_forward(x, w, b)
correct_out = np.array([[ 1.49834967,  1.70660132,  1.91485297],
                        [ 3.25553199,  3.5141327,   3.77273342]])

# Compare your output with ours. The error should be around e-9 or less.
print('Testing affine_forward function:')
print('difference: ', rel_error(out, correct_out))
````

## Answer:
由于Sigmoid函数在函数的尾部分（即输入值非常大的正负值处）的梯度接近于零，因此它遭受梯度消失问题。在一维情况下，可能会出现饱和现象，例如，考虑非常大的正负值[-1e3，1e3]。

ReLU相对于Sigmoid的最大优势之一是它对正输入的线性响应使其不容易遭受梯度消失问题。ReLU的梯度为0（对于负输入）或1（对于正输入）。ReLU可能会遭受梯度消失问题，当所有输入值都为负数时，这不是一个非常可能的情况。当这种情况发生时，一些神经元无法继续训练。这被称为“死ReLU问题”。在一维情况下，可能会出现梯度消失问题，例如，只考虑负值[-1，-2，-3]。

Leaky ReLU通过在负值上应用小的负斜率来解决ReLU中“死亡”神经元的问题，即，如果x < 0，则为0.01x，否则为x。因此，Leaky ReLU旨在解决梯度消失问题。但是，由于max（0.01x，x）函数在x = 0处不连续，因此x = 0处的梯度未定义。因此，如果代码中没有明确处理，可能会出现梯度为零的一维例子，例如，考虑所有零值[0，0，0]，这只会发生在糟糕的网络初始化情况下。

# Solver
Open the file `cs231n/solver.py` and read through it to familiarize yourself with the API. After doing so, use a `Solver` instance to train a `TwoLayerNet` that achieves about `36%` accuracy on the validation set.

````python
input_size = 32 * 32 * 3
hidden_size = 50
num_classes = 10
model = TwoLayerNet(input_size, hidden_size, num_classes)

solver = Solver(model, data,
                update_rule='sgd',
                optim_config={
                    'learning_rate': 1e-3,
                },
                lr_decay=0.95,
                num_epochs=10, batch_size=100,
                print_every=100)
solver.train()
scores = model.loss(data['X_test'])
y_pred = np.argmax(scores, axis = 1)
acc = np.mean(y_pred == data['y_test'])
print('test acc: %f'%acc)
````
